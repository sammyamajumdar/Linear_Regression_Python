```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv("desktop/Neural_Network/ex1data1.txt", header = None)
print(data)
X = data[0]
y = data[1]
# visualize the data
plt.scatter(X, y, marker = "x", color = "r")
plt.xlabel("population in 10,000s")
plt.ylabel("profits in $10,000s")
plt.show()
```

```python
# adding the intercept term
X = X[:, np.newaxis]
y = y[:, np.newaxis]
theta = np.zeros([2,1])
iterations = 1500
learning_rate = 0.01 #alpha
ones = np.ones((len(y), 1)) #len(y) is the number of training examples
X = np.hstack((ones, X))
```

```python
def compute_cost(X, y, theta):
    temp = np.dot(X, theta) - y
    return np.sum(np.power(temp,2))/(2 * len(y))
J = compute_cost(X, y, theta)
print(J)
```

```python
def gradient_descent(X, y, theta, iterations, learning_rate):
    for i in range(iterations):
        temp = np.dot(X, theta) - y
        temp = np.dot(X.T, temp)
        theta = theta - (learning_rate/len(y))*temp
    return theta
theta = gradient_descent(X, y, theta, iterations, learning_rate)
print(theta)
```

```python
print(compute_cost(X, y, theta))
```
