```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("desktop/DL_ML/ex1data2.txt", header = None)
X = data.iloc[:, 0:2] # the first 2 columns are X values
y = data.iloc[:, 2] # the third column are Y values 
m = len(y) # number of training examples
print(data.head())
print(data.describe())
```

```python
plt.scatter(X[0],y, marker = "x", color = "r", label = "size")
plt.scatter(X[1],y, marker = "x", color = "b", label = "bedrooms")
plt.xlabel("Size of House in Square Feet/Bedrooms")
plt.ylabel("Housing Prices in $100,000s")
plt.legend()
plt.show()
```

```python
# Feature normalisation 
X = (X - np.mean(X))/np.std(X)
# Visualize the scaled data
plt.scatter(X[0],y, marker = "x", color = "r", label = "size")
plt.scatter(X[1],y, marker = "x", color = "b", label = "bedrooms")
plt.xlabel("Size of House in Square Feet/Bedrooms")
plt.ylabel("Housing Prices in $100,000s")
plt.legend()
plt.show()
```

```python
# Adding the Intercept Term
ones = np.ones((m,1))
X = np.hstack((ones, X))
iterations = 500
learning_rate = 0.01
theta = np.zeros((3, 1))
y = y[:, np.newaxis]
```

```python
def compute_cost(X, y, theta):
    temp = np.dot(X, theta) - y
    return np.sum(np.power(temp, 2)) / (2*m)
J = compute_cost(X, y, theta)
print("the cost is : " + str(J))
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    for i in range(iterations):
        temp = np.dot(X, theta) - y
        temp = np.dot(X.T, temp)
        theta = theta - (alpha/m) * temp
    return theta
theta, iterations_plot, theta_plot = gradient_descent(X, y, theta, learning_rate, iterations)
print("the optimised theta values are : " + str(theta))
```

```python
print("the new cost with optimised values of theta is : " + str(compute_cost(X, y, theta)))
```
